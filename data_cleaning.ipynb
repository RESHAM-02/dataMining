```python
#@import dataset using pandas library
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Customer Call List.xlsx')
df
```

```python
df =df.drop_duplicates()
df
```

```python
df= df.drop(columns='Not_Useful_Column')
df
```

```python
#@ removing unnecessary characters from last name
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# prompt: select Phone_Number columns and make phone number as in the format of : 3 digits-3 digits- 4 digits. Also consider the numbers with no special characters and do as formatted to them

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])

df
```

```python
# prompt: The numbers like 7066950392 are displayed as nan--. Fix them as well

# Function to fix 'nan--' numbers
def fix_nan_numbers(phone):
    if isinstance(phone, str) and phone.lower() == 'nan--':
        return None  # Or return an empty string, depending on desired output
    return phone

df["Phone_Number"] = df["Phone_Number"].apply(fix_nan_numbers)
df
```

```python
#@ Parse Address column into Street_Address , State and Zip_Code
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',expand=True)
df
```

```python
#@ lets make Paying customer to be Y for Yes and N for No
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
#@ lets make Do_Not_Contact to be Y for yes and N for no
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
#@ Let's replace The n/a and NaN from Do_Not_Contact columns with empty strings
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna('')
df
```

```python
#@ Let's do the same with Paying Customers column
df["Paying Customer"] = df["Paying Customer"].fillna('')
df
```

```python
#@ Replacing N/a and NaN with empty string
df["Phone_Number"] = df["Phone_Number"].fillna('')
df
```

```python
#@ Let's replace Non-e- and Na-- with empty string in Phone_Number column
df["Phone_Number"] = df["Phone_Number"].str.replace('Non-e-','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df

```

```python
#@ Let's drop the rows where Do_Not_Contact column are Y
df = df[df["Do_Not_Contact"] != 'Y']
df
```

```python
# prompt: #@ drop the rows with blank Phone_Number

df = df[df['Phone_Number'] != '']
df
```

```python
# prompt: reset serial no from 1

df = df.reset_index(drop=True)
df
```
